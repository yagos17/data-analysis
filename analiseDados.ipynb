Objetivo: Realizar uma análise simples, porém útil, de vendas, incluindo cálculo de vendas totais, vendas por produto e loja, e controle de estoque restante.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Vendas.xlsx')

df
```

**Valor Total das Vendas**

```python
df['Valor Total'] = df['Preço'] * df['Quantidade Vendida']
df
```

**Total de Vendas por Produto**

```python
vendas_por_produto = df.groupby('Produto')['Quantidade Vendida'].sum().reset_index()
vendas_por_produto
```

**Total de Vendas por Loja**

```python
vendas_por_loja = df.groupby('Loja')['Quantidade Vendida'].sum().reset_index()
vendas_por_loja
```

**Estoque Restante**

```python
df['Estoque Restante'] = df['Estoque Inicial'] - df['Quantidade Vendida']
df
```

```python
df.to_excel('/content/drive/MyDrive/Vendas.xlsx', index=False)
```
